The original text file of the data dictionary used to construct the "Match" table in the database [can be found online here](http://www.football-data.co.uk/notes.txt).

**The newly created data dictionary for this table can be found [here](file:./datsets_database/soccer_data_dictionary.md).**

```python
# Import needed libraries
import pandas as pd
import sqlite3 as sql

pd.options.display.max_columns = 115
```

I built a function that queries the database and returns a Pandas dataframe based on the table input of the user's choice.

```python
# create the connection to the database file and create a dataframe for the "Match Table"
def create_table_df(table):
    db = "./database.sqlite"
    connection = sql.connect(db)

    query = '''SELECT * FROM {}'''.format(table)
    df = pd.read_sql_query(query, connection)
    
    return df
```

```python
match_df = create_table_df("Match")
match_df.head()
```
